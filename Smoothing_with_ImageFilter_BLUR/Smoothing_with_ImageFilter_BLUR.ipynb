```python
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

def add_salt_and_pepper_noise(image, prop_noise):
    # Calculate the number of pixels to add noise to, based on the specified proportion
    n = int(image.width * image.height * prop_noise)

    # Generate 'n' random x and y coordinates within the image's dimensions
    x, y = np.random.randint(0, image.width, n), np.random.randint(0, image.height, n)

    # Iterate over the randomly generated x and y coordinates simultaneously using zip
    for (x, y) in zip(x, y):
        # For each (x, y) coordinate pair, add salt-and-pepper noise to the image
        # with a probability of 50% of turning the pixel completely black (0,0,0)
        # and 50% of turning the pixel completely white (255, 255, 255)
        image.putpixel((x, y), ((0, 0, 0) if np.random.rand() < 0.5 else (255, 255, 255)))

    return image

def plot_image(image, title):
    plt.imshow(image, extent=[0, 1, 0, 1])  # Set extent parameter for consistent size
    plt.title(title)
    plt.axis('off')
    plt.show()

def main():
    # Load the original image of the mandrill using the Image module from Pillow
    image_path = '/content/drive/MyDrive/Datasets/33-mi/test/Callisia fragrans/103.JPG'
    original_image = Image.open(image_path)

    # Add salt-and-pepper noise to the image with a specified proportion
    prop_noise = 0.1  # You can adjust this value for the desired noise level
    noisy_image = add_salt_and_pepper_noise(original_image.copy(), prop_noise)

    # Apply a blur filter to the noisy image using the ImageFilter module from Pillow
    blurred_image = noisy_image.filter(ImageFilter.BLUR)

    # Display the original image with added noise
    plot_image(noisy_image, f'Original Image with {int(100 * prop_noise)}% added noise')

    # Display the smoothed image after applying the blur filter
    plot_image(blurred_image, 'Smoothing with ImageFilter.BLUR')

if __name__ == "__main__":
    main()
```
