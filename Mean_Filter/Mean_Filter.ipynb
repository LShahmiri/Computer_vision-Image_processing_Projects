```python
import cv2
import numpy as np
import scipy.ndimage
from google.colab.patches import cv2_imshow
# opening image from cv2
Image_ = cv2.imread('/content/drive/MyDrive/Datasets/33-mi/test/Angelica dahurica/10.JPG')

# converting image to grayscale
Image_Gray = cv2.cvtColor(Image_, cv2.COLOR_BGR2GRAY)

# Initializing  filter 5*5
# it is divided to 25 for normalization
k = np.ones((5, 5))/25

# performing convolution
b = scipy.ndimage.filters.convolve(Image_Gray, k)

# writing the image to a file
cv2.imwrite('/content/drive/MyDrive/Datasets/mean_output_image.jpg', b)

# showing all process
Image_mean = cv2.imread('/content/drive/MyDrive/Datasets/mean_output_image.jpg')
cv2_imshow(Image_)
cv2_imshow(Image_Gray)
cv2_imshow(Image_mean)



```
