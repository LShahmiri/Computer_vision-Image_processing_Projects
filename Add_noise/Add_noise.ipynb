```python
#adding some random-samples based on some distribution.
import numpy as np
import cv2
import numpy as np
img=cv2.imread("/content/drive/MyDrive/dif-k/pure/Moringa oleifera/65.JPG")
mean = 5.0  
std =5.0    
noisy_img = img + np.random.normal(mean, std, img.shape)
noisy_img_clipped = np.clip(noisy_img, 0, 255) 
```

```python
from google.colab.patches import cv2_imshow
#noisy image
cv2_imshow(noisy_img_clipped)
#pure image
cv2_imshow(img)
```
